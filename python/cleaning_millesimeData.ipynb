```python
import pandas as pd
import os
import csv
```

```python
# Loading the translation file
with open('Resources/translate.csv', newline='', encoding="utf-8") as csvfile:
    translation_file = csv.reader(csvfile, delimiter=',')
    tr_list = []
    for row in translation_file:
        tr_list.append([str(row[0]), str(row[1]), str(row[2])])
```

```python
# Loading, cleaning the vintage files
countries = [ "Italy", "Spain", "USA","France"]
list_df = []
for cc in countries:
    print(cc)
    csv_path = os.path.join("Resources","millesimesTable_"+cc+".csv")
    df = pd.read_csv(csv_path, encoding="utf-8")

    # Add the column country
    df['Country']=cc

    # Rename the column region
    df = df.rename( columns={"RÃ©gion": "Region"})
    
    # Use the translation list to translate the wine and region names
    for tt in tr_list:
        try:
            df[tt[0]] = df[tt[0]].str.replace(tt[1], tt[2])
        except:
            pass

    # rearrange the columns in order to have the country first
    cols = list(df.columns.values) #Make a list of all of the columns in the df
    years_cols = cols[2:-1]
    cols.pop(cols.index('Country')) #Remove b from list
    # cols.pop(cols.index('x')) #Remove x from list
    df = df[['Country']+cols] #Create new dataframe with columns in the order you want

    # Cast the columns of the years into integer or when no values NaN
    # loop on each year column
    for yy in years_cols:
        try:
            df[yy] = df[yy].str.replace('-','')
        except:
            pass
        df[yy] = df[yy].astype(int, errors='ignore')
        
    list_df.append(df)
```

```python
# pd.concat([list_df[0], list_df[1]], sort=False)
merged_dt = pd.concat(list_df, sort=False)
```
