```python
import pandas as pd
import re
import os
import numpy as np
```

```python
# file= "Resources/LIGHT_winemag-data-130k-v2.csv"
# output_data_file = "Resources_clean/WineList_year.csv"
# file_pd= pd.read_csv(file, encoding= "utf-8")

file= "Resources/winemag-data-130k-v2.json"
output_data_file = "Resources/winemag_years.json"
file_pd= pd.read_json(file)
```

```python
file_pd["Wine Year"]= ""
file_pd.head()
```

```python
# replacing the first column with the index
file_pds= file_pd.dropna()
file_pds = file_pds.rename(columns={ file_pds.columns[0]: "index" })
file_pds = file_pds.set_index('index')
file_pds.head()
```

```python
file_pds.describe()
```

```python
for index, row in file_pds.iterrows():
    wine= row['title']
    try:
        wine_year= re.findall(r'\b\d+\b', wine)
        if 1000 < int(wine_year[0]) < 2020:
                file_pds.loc[index, "Wine Year"] = wine_year[0]
        elif 1000 < int(wine_year[1]) < 2020:
            file_pds.loc[index, "Wine Year"] = wine_year[1]
        else:
            file_pds.loc[index, "Wine Year"]= ""      
    except (KeyError, IndexError):
        print("Year not found. Skipping...")
```

```python
csv_pd= file_pds.replace('', np.nan, inplace=True)
csv_pd= file_pds.dropna()
csv_pd.head()
```

```python
csv_pd.to_csv(output_data_file)
```
